```python
rodutos_vendidos = []

while True:
  produto = input("Digite o nome do produto (ou pressione Enter para finalizar): ")
  if not produto:
    break

  try:
    quantidade = int(input("Digite a quantidade vendida: "))
  except ValueError:
    print("Quantidade inválida. Digite um número inteiro.")
    continue

  produtos_vendidos.append((produto, quantidade))

print("\nProdutos vendidos:")
for produto, quantidade in produtos_vendidos:
  print(f"{produto}: {quantidade}")
```

```python

import random

while True:
  jogar = input('Digite "jogar" para lançar o dado ou "sair" para sair: ')
  if jogar.lower() == 'sair':
    break
  elif jogar.lower() == 'jogar':
    # Simula um dado viciado, com maior probabilidade de sair 6
    # 70% de chance de ser 6, 30% dividido entre os outros números
    if random.random() < 0.7:
      resultado = 6
    else:
      resultado = random.randint(1, 5)
    print("Resultado do dado:", resultado)
  else:
    print('Comando inválido. Digite "jogar" ou "sair".')
```

### **3 - Quiz Interativo**
Descrição:
Monte um pequeno quiz com perguntas e múltiplas alternativas.

Regras:

```python


import random

def quiz():
    perguntas = [
        {
            "pergunta": "Qual o teorema fundamental da álgebra?",
            "opcoes": ["a) Teorema de Pitágoras", "b) Teorema de Fermat", "c) Todo polinômio de grau n tem n raízes complexas"],
            "resposta": "c"
        },
        {
            "pergunta": "Quem desenvolveu a teoria da relatividade geral?",
            "opcoes": ["a) Isaac Newton", "b) Albert Einstein", "c) Niels Bohr"],
            "resposta": "b"
        },
        {
            "pergunta": "O que é o princípio da incerteza de Heisenberg?",
            "opcoes": ["a) Não é possível determinar simultaneamente a posição e a velocidade de uma partícula", "b) A energia é sempre constante", "c) A luz pode ser dividida em partículas"],
            "resposta": "a"
        },
        {
            "pergunta": "Qual é o nome da partícula subatômica que foi descoberta no acelerador de partículas Large Hadron Collider?",
            "opcoes": ["a) Bóson de Higgs", "b) Elétron", "c) Neutrino"],
            "resposta": "a"
        },
        {
            "pergunta": "Em que ano o Império Romano do Ocidente foi oficialmente dissolvido?",
            "opcoes": ["a) 395 d.C.", "b) 476 d.C.", "c) 509 a.C."],
            "resposta": "b"
        },
        {
            "pergunta": "Qual é o nome da teoria que descreve a origem do universo a partir de um ponto de densidade infinita?",
            "opcoes": ["a) Teoria do Big Bang", "b) Teoria da evolução cósmica", "c) Teoria do multiverso"],
            "resposta": "a"
        },
        {
            "pergunta": "Quem foi o primeiro filósofo a propor o conceito de 'ápeiron', o princípio fundamental de todas as coisas?",
            "opcoes": ["a) Heráclito", "b) Parmênides", "c) Anaximandro"],
            "resposta": "c"
        },
        {
            "pergunta": "Qual foi a principal causa da Revolução Russa de 1917?",
            "opcoes": ["a) Crise econômica e insatisfação popular", "b) Ascensão do fascismo", "c) Revoltas militares contra o czarismo"],
            "resposta": "a"
        },
        {
            "pergunta": "O que é a 'matéria escura' no contexto da física moderna?",
            "opcoes": ["a) Uma forma de matéria que não interage com a luz", "b) Uma substância conhecida que é invisível", "c) A matéria que compõe os buracos negros"],
            "resposta": "a"
        }
    ]

    random.shuffle(perguntas)  # Embaralha as perguntas
    num_perguntas = random.randint(3, 5)  # Define um número aleatório de perguntas
    acertos = 0

    for i in range(num_perguntas):
        print(f"\nPergunta {i+1}: {perguntas[i]['pergunta']}")
        for opcao in perguntas[i]['opcoes']:
            print(opcao)
        resposta_usuario = input("Sua resposta: ").lower()
        if resposta_usuario == perguntas[i]['resposta']:
            print("Correto!")
            acertos += 1
        else:
            print(f"Errado. A resposta correta é {perguntas[i]['resposta']}.")

    print(f"\nVocê acertou {acertos} de {num_perguntas} perguntas.")

quiz()

```

A ideia é criar um jogo totalmente jogável pelo terminal, onde o jogador deve adivinhar uma palavra secreta chutando letras, uma de cada vez. Se errar demais, perde. Se acertar tudo, vence!

```python
import random

def jogar_forca():
    palavra_secreta = "python"  # Palavra secreta (pode ser modificada)
    letras_certas = ["_"] * len(palavra_secreta)
    tentativas = 6
    letras_erradas = []

    print("Bem-vindo ao Jogo da Forca!")

    while tentativas > 0 and "_" in letras_certas:
        print("\nPalavra:", " ".join(letras_certas))
        print("Tentativas restantes:", tentativas)
        print("Letras erradas:", ", ".join(letras_erradas))

        chute = input("Digite uma letra: ").lower()

        if len(chute) != 1 or not chute.isalpha():
            print("Chute inválido. Digite uma única letra.")
            continue

        if chute in letras_certas or chute in letras_erradas:
            print("Você já tentou essa letra. Tente outra.")
            continue

        if chute in palavra_secreta:
            for i in range(len(palavra_secreta)):
                if palavra_secreta[i] == chute:
                    letras_certas[i] = chute
        else:
            tentativas -= 1
            letras_erradas.append(chute)
            print("Letra incorreta.")

    if "_" not in letras_certas:
        print("\nParabéns! Você venceu! A palavra era:", palavra_secreta)
    else:
        print("\nVocê perdeu! A palavra era:", palavra_secreta)

jogar_forca()
```
