```python
# Agora, imagime que são muitos e muitos produtos. O que voce faria?
# usando um for combinado com uma variável "incremental", eu só
# escrevo o comando print uma vez e ele rodará até que pecorra toda a lista

produtos = ["banana" , "maçã" , "pera" , "uva", "abacaxi" , "manga", "mouse" , "pá"]
vendas = [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 100 ]

for i in range( len(produtos) ):
    print(f"A venda do produto { produtos[i] } foi de { vendas [i] } \n" + "-"*30)
```

```python
# Eu posso combinar o FOR com IF , para checar se cada elemento de uma lista
# atende - ou não - a uma dada condição
lista_vendas = [100,150,300,500,660,250,140,50]
meta = 160

for venda in lista_vendas:
    if(venda >= 160):
        print(venda , "- BATEU A META")
    else:
        print(venda , "- NÃO BATEU A META")

```

```python
# DESAFIO: Incremente o código abaixo para que, ao final, ele diga qual a porcentagem de vendedores
# que bateram a meta.
# O codigo deve funcionar , mesmo que sejam adicionados novas vendas a lista, ou se meta mudar.

#resolução:
#preciso descobrir o total de vendedores e q qtd que bateu a meta
lista_vendas = [100,150,300,500,660,250,140,50]
meta = 160
total_vendedores = len(lista_vendas)
qtd_bateram_meta = 0

for venda in lista_vendas:
    if(venda >= meta):
        print(venda , "- BATEU A META")
        qtd_bateram_meta += 1 #contador de vendedores que bateram meta
    else:
        print(venda , "- NÃO BATEU A META")

porcento_bateram_meta = qtd_bateram_meta / total_vendedores

print(f"Um total de { qtd_bateram_meta } vendedores bateram a meta")
print(f"Isso corresponde a {porcento_bateram_meta:.1%} do total de vendedores" )
```

### Enumerate - Uma função muito util que traz uma "lista" onde cada elemento contem o indice e o valor de uma lista original

```python
#Exemplo
lista = ["banana" , "maça" , "uva" ]

objeto_enumerate = enumerate(lista)  # o enumerate ira retornar um objeto, contendo indices e valores da lista original

lista_final = list(objeto_enumerate) # o comando list() converte o objeto gerado para uma nova lista

print(lista_final) # imprimimos a lista gerada, onde cada elemento é uma tupla contendo indice+valor dos elementos)
```

```python
for elemento in lista_final:
    print (elemento)  # aqui ele vai imprimir cada elemento, que é uma tupla de 2 dados
```

```python
# UNPACKING - TRAZ CADA ELEMENTO PARA UMA VARIAVEL
# ATENÇÃO : OBRIGATORIO TER EXATAMENTE O MESMO NUMERO DE VARIAVEIS E DE ELEMENTOS
tupla = ("anderson" , "28" , "34857434")
nome , idade , cpf = tupla

print (nome , idade , cpf)
```

```python
# Agora vamos usar tudo junto: Enumerate / Unpacking e FOR

produtos = ["banana" , "maçã" , "pera" , "uva", "abacaxi" , "manga", "mouse" , "pá"]
vendas = [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 100 ]

#Exemplo 1: usando apenas a variavel produto , ela irá carregar todo o elemento, que uma tupla
for produto in enumerate(produtos):
    print(produto)
```

```python
# Exemplo 2: usando o unpacking, o elemento "da vez" é quebrado em duas variaveis:
# indice e produto

for indice,produto in enumerate(produtos):
    print("O produto no indice", indice , "é o" , produto)
```

```python
# Exemplo 3: Agora vamos usar tudo junto e ainda por cima usa ro indice para
# acessar o elemento na lista vendas

produtos = ["banana" , "maçã" , "pera" , "uva", "mochila" , "manga", "tela" , "pá"]
vendas = [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 100 ]

for indice,produto in enumerate(produtos):
    print("O produto no indice", indice , "é a" , produto)
    print("Este produto vendeu",vendas[indice])
    print("-" * 30)
```

```python
# Exemplo 4: Agora combinar com o IF
# DESAFIO-RELAMPAGO: faça imprimir mais uma linha informando se a venda do produto
# bateu a meta

meta = 250
produtos = ["banana" , "maçã" , "pera" , "uva", "mochila" , "manga", "tela" , "pá"]
vendas = [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 100 ]

for indice,produto in enumerate(produtos):
    print("O produto no indice", indice , "é a" , produto)
    print("Este produto vendeu",vendas[indice])
    if(vendas[indice] >= meta) :
        print("Bateu a meta")
    else:
        print("Não bateu a meta")
    print("-" * 30)
```

```python
# Exemplo 5: Quando eu tenho uma situação onde há um iterável dentro de outro,
# (p. ex. lista dentro de lista, tuplas dentro de lista, etc.etc), e eu preciso percorrer cada elemento,
# a solução é fazer um for dentro de um for

estoque_minimo = 250
produtos = ["banana" , "maçã" , "pera" , "uva", "mochila" , "manga", "tela" , "pá","banana" , "maçã" , "pera" , "uva", "mochila" , "manga", "tela" , "pá",]
estoque_fabricas = [
                    [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 300, 100 , 200 , 300, 1000 , 500 , 250 , 540 , 300 ],
                    [ 80 , 50 , 300, 500 , 400 , 250 , 540 , 100, 80 , 50 , 300, 500 , 400 , 250 , 540 , 100 ],
                    [ 100 , 200 , 300, 1000 , 500 , 250 , 540 , 300, 100 , 200 , 300, 1000 , 500 , 250 , 540 , 300 ],
                    [ 80 , 50 , 300, 1000 , 400 , 100 , 540 , 100,80 , 50 , 300, 1000 , 400 , 100 , 540 , 100  ],
                    [ 100 , 40 , 300, 200 , 500 , 250 , 540 , 300,100 , 40 , 300, 200 , 500 , 250 , 540 , 300 ],
                    [ 80 , 50 , 300, 1000 , 400 , 250 , 540 , 100,80 , 50 , 300, 1000 , 400 , 250 , 540 , 100 ]
]

# DESAFIO:
# O programa deverá imprimir o nome do produto, estoque da farica 1 e se está OK ou abaixo do nivel de segurança,
# estoque da farica 1 e se está OK ou abaixo do nivel de segurança,
# EXEMPLO DA SAÍDA DESEJADA:
# Produto: Banana
# Estoque Fábrica 1: 100 - estoque baixo, repor mercadoria
# Estoque Fábrica 2: 80 - estoque baixo, repor mercadoria
# ------------------------

for indiceproduto,produto in enumerate(produtos):
    print("O produto no indice", indiceproduto , "é a" , produto)
    for indicefabrica, fabrica in enumerate(estoque_fabricas):
        if(fabrica[indiceproduto] >= estoque_minimo):
            status = "OK"
        else:
            status = "Estoque Baixo, repor mercadoria"
        print(f"Estoque Fábrica {indicefabrica+1}: {fabrica[indiceproduto]} - {status} ")
    print("-" * 30)



```
