🔍 O que o programa faz?

Ele codifica 3 palavras trocando cada letra pela próxima na tabela ASCII (ex: a vira b, b vira c, etc).

Depois, ele decodifica uma palavra, fazendo o caminho inverso — ou seja, volta uma letra na tabela ASCII (ex: b vira a).

✅ Regras importantes:

Usa ord() para pegar o número da letra.

Usa chr() para transformar número de volta em letra.

Usa acesso por índice (palavra[0], palavra[1], etc).

Não usa laços (for ou while) — tudo feito "na mão".

✨ Exemplo:

Se você digitar:

hello → vira ifmmp

world → vira xpsme

agent → vira bhfou

Depois, se você digitar ifmmp, ele mostra: hello.



```python

# Pedindo as três palavras secretas
palavra1 = input("Digite a primeira palavra secreta (até 5 letras): ")
palavra2 = input("Digite a segunda palavra secreta (até 5 letras): ")
palavra3 = input("Digite a terceira palavra secreta (até 5 letras): ")

# Codificando a primeira palavra
letra1_1 = chr(ord(palavra1[0]) + 1) if len(palavra1) > 0 else ''
letra1_2 = chr(ord(palavra1[1]) + 1) if len(palavra1) > 1 else ''
letra1_3 = chr(ord(palavra1[2]) + 1) if len(palavra1) > 2 else ''
letra1_4 = chr(ord(palavra1[3]) + 1) if len(palavra1) > 3 else ''
letra1_5 = chr(ord(palavra1[4]) + 1) if len(palavra1) > 4 else ''
palavra1_codificada = letra1_1 + letra1_2 + letra1_3 + letra1_4 + letra1_5

# Codificando a segunda palavra
letra2_1 = chr(ord(palavra2[0]) + 1) if len(palavra2) > 0 else ''
letra2_2 = chr(ord(palavra2[1]) + 1) if len(palavra2) > 1 else ''
letra2_3 = chr(ord(palavra2[2]) + 1) if len(palavra2) > 2 else ''
letra2_4 = chr(ord(palavra2[3]) + 1) if len(palavra2) > 3 else ''
letra2_5 = chr(ord(palavra2[4]) + 1) if len(palavra2) > 4 else ''
palavra2_codificada = letra2_1 + letra2_2 + letra2_3 + letra2_4 + letra2_5

# Codificando a terceira palavra
letra3_1 = chr(ord(palavra3[0]) + 1) if len(palavra3) > 0 else ''
letra3_2 = chr(ord(palavra3[1]) + 1) if len(palavra3) > 1 else ''
letra3_3 = chr(ord(palavra3[2]) + 1) if len(palavra3) > 2 else ''
letra3_4 = chr(ord(palavra3[3]) + 1) if len(palavra3) > 3 else ''
letra3_5 = chr(ord(palavra3[4]) + 1) if len(palavra3) > 4 else ''
palavra3_codificada = letra3_1 + letra3_2 + letra3_3 + letra3_4 + letra3_5

# Exibindo a mensagem codificada
mensagem_codificada = f"A mensagem secreta é: {palavra1_codificada} {palavra2_codificada} {palavra3_codificada}"
print(mensagem_codificada)


# Desafio extra: decodificando uma palavra
palavra_codificada = input("Digite uma palavra codificada (até 5 letras): ")

# Decodificando a palavra
letra_dec_1 = chr(ord(palavra_codificada[0]) - 1) if len(palavra_codificada) > 0 else ''
letra_dec_2 = chr(ord(palavra_codificada[1]) - 1) if len(palavra_codificada) > 1 else ''
letra_dec_3 = chr(ord(palavra_codificada[2]) - 1) if len(palavra_codificada) > 2 else ''
letra_dec_4 = chr(ord(palavra_codificada[3]) - 1) if len(palavra_codificada) > 3 else ''
letra_dec_5 = chr(ord(palavra_codificada[4]) - 1) if len(palavra_codificada) > 4 else ''

palavra_decodificada = letra_dec_1 + letra_dec_2 + letra_dec_3 + letra_dec_4 + letra_dec_5
print(f"Palavra decodificada: {palavra_decodificada}")
```
