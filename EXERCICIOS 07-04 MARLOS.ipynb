## 1. Criando um Registro de HÃ³spedes

Digamos que vocÃª estÃ¡ criando o sistema para registrar a chegada de hÃ³spedes em um hotel. No hotel, os hÃ³spedes podem ter quartos com 1, 2, 3 e 4 pessoas. Seu sistema deve conseguir:

1. Identificar quantas pessoas o hÃ³spede que acabou de chegar vai ter no quarto (perguntando por meio de input)
2. De acordo com a quantidade de pessoas do hÃ³spede, ele deve fazer um for para perguntar o cpf e o nome de cada pessoa, a fim de registrÃ¡-la no quarto (2 inputs para cada pessoa, 1 para o cpf e outro para o nome)
3. O seu programa entÃ£o deve gerar uma lista com todas as pessoas que ficarÃ£o no quarto em que cada item dessa lista Ã© o nome da pessoa e o cpf da pessoa, assim:

- Para simplificar, nÃ£o vamos nos preocupar com possibilidades de "tentar colocar mais de 1 hÃ³spede, digitar o cpf errado, etc. Nosso objetivo Ã© treinar a criaÃ§Ã£o de uma rotina de cadastro

```python

qtde_pessoas = int(input('Quantas pessoas terÃ£o no quarto? '))

quarto = []

for i in range(qtde_pessoas):
  nome = input(f'Digite o nome da {i+1}Âª pessoa: ')
  cpf = input(f'Digite o CPF da {i+1}Âª pessoa: ')
  quarto.append([nome, cpf])

quarto
```

4. Contando a Grana da Feira

[2.50, 3.00, 1.75, 4.20, 3.30]

Desafio: Crie um cÃ³digo que use for para somar o total gasto na feira.





```python
# prompt: Contando a Grana da Feira SituaÃ§Ã£o: VocÃª foi na feira comprar frutas. Anote os preÃ§os das frutas que comprou: [2.50, 3.00, 1.75, 4.20, 3.30]
# Desafio: Crie um cÃ³digo que use for para somar o total gasto na feira.

precos = [2.50, 3.00, 1.75, 4.20, 3.30]
total = 0
for preco in precos:
  total = total + preco
print(f'O total gasto na feira foi de R$ {total}')
```

9. Contando Vogais de uma Palavra
SituaÃ§Ã£o: VocÃª estÃ¡ criando um jogo de palavras e precisa contar as vogais de uma palavra.

Desafio: PeÃ§a uma palavra e use for para contar quantas vogais ela tem.

```python
# prompt: Contando Vogais de uma Palavra SituaÃ§Ã£o: VocÃª estÃ¡ criando um jogo de palavras e precisa contar as vogais de uma palavra.
# Desafio: PeÃ§a uma palavra e use for para contar quantas vogais ela tem. palavra = input("Digite uma palavra: ").lower()

palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
total_vogais = 0

for letra in palavra:
  if letra in vogais:
    total_vogais += 1

print(f"A palavra '{palavra}' tem {total_vogais} vogais.")
```
